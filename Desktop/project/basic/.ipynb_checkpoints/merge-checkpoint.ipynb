```python
# 데이터 불러오기
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/mergeTEst.csv',index_col= 0)
df.head()
```

## Q1.df1과 df2 데이터를 하나의 데이터 프레임으로 합쳐라

```python
# data
df1 = df.iloc[:,:4]
df2 = df.iloc[:,4:]

display(df1)
display(df2)
```

```python
# 풀이
res_df = pd.merge(df1, df2, on="Location")
res_df
```

```python
#결과
res_df
```

```python
# data
df3 = df.iloc[:2,:4].T
df3.index.names = ['year']

df4 = df.iloc[5:,3:].T
df4.index.names = ['year']

display(df3)
display(df4)
```
