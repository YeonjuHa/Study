# Matplotlib

## 막대그래프 그리기

```python
import matplotlib.pyplot as plt

x = ['2021', '2022', '2023']
y = [100, 200, 300]

plt.bar(x, y)
```

```python
import matplotlib.pyplot as plt

x = ['cat', 'dog', 'fish', 'bird', 'snake', 'bee']
y = [98, 84, 45, 55, 68, 72]

plt.bar(x, y)
```

```python
# seaborn 형식
import seaborn as sns

x = ['cat', 'dog', 'fish', 'bird', 'snake', 'bee']
y = [98, 84, 45, 55, 68, 72]

sns.barplot(x=x, y=y)
```

```python
import matplotlib.pyplot as plt

x = range(6)
y = range(10, 70, 10)

plt.bar(x, y, width=1, align='edge', color='red', edgecolor='yellow', linewidth=2)
```

## figure 객체

```python
import matplotlib.pyplot as plt

x = range(6)
y = range(10, 70, 10)

plt.bar(x, y, width=1, align='edge', color='red', edgecolor='yellow', linewidth=2)

plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales($)')
plt.legend(['sales'])
plt.grid(True, axis='y')
```

## 서브플롯 (subplots)

```python
import matplotlib.pyplot as plt

fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(nrows=2, ncols=3)
fig.set_size_inches(16, 8)

ax1.set(title = '1st plot')
ax2.set(title = '2nd plot')
ax3.set(title = '3rd plot')
ax4.set(title = '4th plot')
ax5.set(title = '5th plot')
ax6.set(title = '6th plot')
```

```python
import matplotlib.pyplot as plt

fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(nrows=2, ncols=3)
fig.set_size_inches(16, 8)

ax1.set(title = '1st plot')
ax2.set(title = '2nd plot')
ax3.set(title = '3rd plot')
ax4.set(title = '4th plot')
ax5.set(title = '5th plot', xlabel='month', ylabel='sales')
ax6.set(title = '6th plot')

# 그래프 추가
x = range(6)
y = range(10, 70, 10)

ax5.bar(x, y, width=1, align='edge', color='red', edgecolor='yellow', linewidth=2)
ax5.grid(True, axis='y')
```

```python
import matplotlib.pyplot as plt

x = range(1, 13)
y = range(10, 130, 10)

fig = plt.figure(figsize=(6.4, 4.8))
plt.bar(x, y)
plt.title('월별 매출')
plt.xlabel('월')
plt.ylabel('매출')
```

```python
import matplotlib.pyplot as plt

x = range(1, 13)
y = range(10, 130, 10)

fig = plt.figure(figsize=(6.4, 4.8))
plt.bar(x, y)

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.title('월별 매출')
plt.xlabel('월')
plt.ylabel('매출')
```

## 음수(-) 표시

```python
import matplotlib.pyplot as plt

x = ['cat', 'dog', 'fish']
y = [30, -40, 20]

plt.bar(x, y)
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

x = ['cat', 'dog', 'fish']
y = [30, -40, 20]

mpl.rcParams['axes.unicode_minus'] = False

plt.bar(x, y)
```

# Seaborn + Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = sns.load_dataset('tips')

fig, ax = plt.subplots(nrows=2, ncols=2)
fig.set_size_inches(16, 8)

sns.scatterplot(x = df['total_bill'], y = df['tip'], ax = ax[0][0])
sns.barplot(x = df['size'], y = df['tip'], ax = ax[0][1])
sns.regplot(x = df['total_bill'], y = df['tip'], ax = ax[1][0])
sns.violinplot(x = df['size'], y = df['tip'], ax = ax[1][1])
```
